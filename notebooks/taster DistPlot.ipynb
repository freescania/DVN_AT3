```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.figure_factory as ff
```

```python
path = "/Users/stefan/UTS/36104 DVN/DVN_AT3/data/"
data = pd.read_csv(path + "winemag-data-130k-v2.csv")
```

```python
data['taster_name'] = data['taster_name'].fillna('Unknown Taster') # fill in unknown tasters

taster_list = data['taster_name'].unique()
taster_list
```

```python
taster_name = "Christina Pickard"

taster = data[(data.taster_name == taster_name)]

#taster
#taster.to_csv(path + "taster.csv")
```

```python
# create new df for points by country
countrypoints = pd.DataFrame()

for i in taster['country']:
    countrypoints[i] = taster[(taster.country == i)].drop(taster.columns.difference(['points']), 1).iloc[:,0].reset_index(drop=True)

    
#drop all columns/countries with < 3 datapoints - needed for distplot
countrypoints = countrypoints.dropna(thresh=3, axis=1) 


# create distribution plot
fig = ff.create_distplot([countrypoints[c].dropna(axis=0) for c in countrypoints.columns],
                         countrypoints.columns,
                         show_hist=False, show_rug=False,
                        #curve_type='normal', # for a normal distribution curve 

                        )

fig.update_layout(
    title_text = "Score Distribution by Country - " + taster_name,
    title_x = 0.5
)

fig.show()

# dropna drops all rows from that columns with nan values
```
